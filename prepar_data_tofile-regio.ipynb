```python
columns_nonnull = []

for index, value in null_capacity.items():
    
    if (value != len(rslt_df.index)):
        print(f"Index : {index}, Value : {value}")
        columns_nonnull.append(index)
print(columns_nonnull)    
```

```python
client_id = 661

rslt_df = pd1.loc[pd1['dataid']==client_id]
pd_t =rslt_df[columns_nonnull]
# print(pd_t)





for i in range(len(columns_nonnull)):
#     print(columns_nonnull[i])
#     print(pd_t[columns_nonnull[i]])
    pd_t[columns_nonnull[i]] = pd_t[columns_nonnull[i]].fillna(method='ffill')
    pd_t[columns_nonnull[i]] = pd_t[columns_nonnull[i]].fillna(method='backfill')    
    
    
print(pd_t)
path = str(client_id) +'.csv'
pd_t.to_csv(path)
print(len(pd_t.index))
```
